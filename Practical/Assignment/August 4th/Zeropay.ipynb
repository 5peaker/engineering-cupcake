```python
import re

# 정규 표현식 패턴 정의
pattern = re.compile(r'^(.*?\S*점업|\S*점)')
service = [] 

# 업종/매장명을 추출하여 출력
for address in addr_k:
    match = pattern.search(address)
    if match:
        print(match.group(1))
        service.append(match.group(1))
```

```python
import re

# addr_k에서 "강동구"가 들어가지 않는 값들을 제거
addr_k = [x for x in addr_k if "강동구" in x]

def extract_address(addr):
    # 주소를 추출하고 전화번호를 제거하는 정규 표현식
    match = re.search(r'(서울특별시 강동구.*?)(?:\s+02|\s+010|\s+070|\s+0|$)', addr)
    if match:
        address = match.group(1).strip()
        # 괄호와 그 안의 내용을 제거하는 정규 표현식
        address = re.sub(r'\s*\(.*?\)', '', address)
        return address
    return None

# 각 변수에서 주소를 추출
extracted_addresses = [extract_address(addr) for addr in addr_k]

# 결과 출력
extracted_addresses
```

```python
def extract_name(addr):
    match = re.search(r'(?:점업|점)\s*(.*?)\s*서울특별시 강동구', addr)
    return match.group(1).strip() if match else None

# 각 변수에서 이름을 추출
extracted_names = [extract_name(addr) for addr in addr_k]

# 결과 출력
extracted_names
```

```python
# extract_address와 extracted_name에서 None을 제거

extracted_addresses = list(filter(None, extracted_addresses))
extracted_names = list(filter(None, extracted_names))
```
